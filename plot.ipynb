```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv("Summary of Announced Launches.csv",';',comment='#',nrows=13,skiprows=[3],index_col=0)
del df['TOTAL']
for year in range(1957,1960):
    df[str(year)] = df['1957-1959']/(3)
for year in range(1960,1970):
    df[str(year)] = df['1960-1969']/(10)
for year in range(1970,1980):
    df[str(year)] = df['1970-1979']/(10)
for year in range(1980,1990):
    df[str(year)] = df['1980-1989']/(10)
del df['1957-1959']
del df['1960-1969']
del df['1970-1979']
del df['1980-1989']
df = df.reindex(sorted(df.columns), axis=1)
df
```

```python

df.loc[['CIS(USSR)','ESA','NASA','US Commercial','TOTAL']].T.plot()
plt.legend()
```
