```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import glob
```

```python
launches_per_year = pd.DataFrame()
for path in glob.glob('launch logs/*.csv'):
    df = pd.read_csv(path,index_col=0,engine='c').sum(axis=0)
    del df['Vehicle']
    df['Year'] = int(path[:-4])
    launches_per_year=launches_per_year.append(df,ignore_index=True)
launches_per_year
```

```python
import matplotlib.pyplot as plt
plt.xlabel('Year')
plt.ylabel('Launches')
plt.plot(launches_per_year['Year'],launches_per_year['Launches'])
```

```python
plt.plot(launches_per_year['Year'],1-launches_per_year['Failures']/launches_per_year['Launches'])
```
