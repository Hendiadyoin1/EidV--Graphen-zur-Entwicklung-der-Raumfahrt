```python
import pandas as pd
import glob
from matplotlib import pyplot as plt
```

```python
counts_all = {}
counts_unique = {}
bodies_per_launch = {}
for file in glob.glob('spaceCraftquerys/*.csv'):
    df = pd.read_csv(file)
    count0 = len(df.axes[0])
    year = int(file[-8:-4])
    if year>=2021:
        continue
    counts_all[year] = count0
    df = df.drop_duplicates(subset=['Launch Date','Site','Vehicle'])
    count1 = len(df.axes[0])
    counts_unique[year] = count1

    bodies_per_launch[year]= float(count0)/float(count1)
```

```python
plt.xlabel('Year')
plt.ylabel('Launches of Rockets and Payloads worldwide')
plt.plot(list(counts_all.keys()),list(counts_all.values()))
# keys = list(counts_all.keys())
# for i in range(1,len(counts_all.keys())-1):
#     if counts_all[keys[i-1]]<counts_all[keys[i]]>counts_all[keys[i+1]]:
#         plt.annotate('(%s, %s)' % (keys[i],counts_all[keys[i]]), xy=(keys[i],counts_all[keys[i]]),size=5)
```

```python
plt.xlabel('Year')
plt.ylabel('Approximate average bodies per Launch')
plt.plot(list(bodies_per_launch.keys()),list(bodies_per_launch.values()))
```
